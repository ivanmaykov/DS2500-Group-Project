```python
import pandas as pd

df = pd.read_csv('notes_freqs.csv')
df = df.dropna()

df.head()
```

```python
# getting unique note names
notes = list(df.loc[:, 'Note Name'].unique())

# dict to hold frequencies and wavelengths of each note
notes_dict = {}

# for every note, finding wavelengths and frequencies in all octaves
for note in notes:
    # lists to store data for each note
    wavelengths = []
    frequencies = []
    
    for idx in range(len(df)):
        if df.loc[idx, 'Note Name'] == note:
            
            # finding and storing wavelengths
            wavelength = df.loc[idx ,'Wavelength (M)*']
            wavelengths.append(wavelength)
            
            # finding and storing frequencies
            freq = df.loc[idx ,'Frequency (Hz)']
            frequencies.append(freq)
            
    # adding all note data to dictionary
    notes_dict[note] = [wavelengths, frequencies]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# getting unique colors for the notes
colors = list(df.loc[:, 'Color'].unique())

sns.set(font_scale=1)

# initializing count to cycle through colors
count = 0

# scatterplotting all notes' frequencies and wavelengths
for key in notes_dict.keys():
    plt.scatter(notes_dict[key][0], notes_dict[key][1], color = colors[count], label = key)
    count += 1
    
# setting title, labels, and legend
plt.suptitle('Wavelength vs. Frequency of Musical Notes')
plt.xlabel('Wavelength')
plt.ylabel('Frequency')
plt.legend()
```

This model shows the relationship between wavelength and frequency and that they are inversely related. It also visualizes the different notes according to their colors, applied to all octaves.
